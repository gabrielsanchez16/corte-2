```python
# Importamos pandas para trabajar tablas y datos
import pandas as pd

# Importamos numpy para crear datos aleatorios
import numpy as np


# Creamos un DataFrame con información de estudiantes
datos = {
    "Nombre": ["Ana", "Luis", "Carlos", "Maria"],
    "Nota1": [4.0, 3.5, 4.5, 5.0],
    "Nota2": [4.2, 3.8, 4.0, 4.8]
}

df = pd.DataFrame(datos)

print("DataFrame inicial")
print(df)


# Sacamos el promedio de las notas
df["Promedio"] = (df["Nota1"] + df["Nota2"]) / 2

print("\nDataFrame con promedio")
print(df)


# Guardamos los datos en un archivo CSV
df.to_csv("estudiantes.csv", index=False)

print("\nArchivo CSV creado correctamente")


# Leemos el archivo CSV
archivo = pd.read_csv("estudiantes.csv")

print("\nContenido del CSV")
print(archivo)


# Creamos datos sintéticos o aleatorios
datos_sinteticos = pd.DataFrame({
    "Edad": np.random.randint(18, 40, 10),
    "Salario": np.random.randint(1000000, 5000000, 10)
})

print("\nDatos sintéticos")
print(datos_sinteticos)


# Mostramos estadísticas básicas
print("\nEstadísticas")
print(datos_sinteticos.describe())
```
