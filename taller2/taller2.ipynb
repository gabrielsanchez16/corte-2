```python
#1. Importo librerías
import pandas as pd
import numpy as np

#2. Creo el DataFrame
ruta = '/content/drive/MyDrive/productos_taller_pandas.xlsx'

df = pd.read_excel(ruta)
df
#Punto 1 – Creación y Exploración del DataFrame

#Primeras 5 filas
df.head()

#Últimas 10 filas
df.tail(10)

#Tipo de dato de cada columna
df.dtypes

#Tamaño del DataFrame
print("Filas y columnas:", df.shape)

#Estadísticas generales
df.describe()

#incluimos la columna categoría:

df.describe(include='all')

#Punto 2 – Operaciones Matemáticas

#Total Venta
df['Total_Venta'] = df['Precio'] * df['Cantidad']

#Costo Total
df['Costo_Total'] = df['Total_Venta'] + df['Costo_Envio']

#Promedio Total Venta
promedio = df['Total_Venta'].mean()

print("Promedio Total Venta:", promedio)

#Producto con mayor Total Venta
mayor = df.loc[df['Total_Venta'].idxmax()]

print(mayor)

#Producto con menor Total Venta
menor = df.loc[df['Total_Venta'].idxmin()]

print(menor)
#Punto 3 – Porcentajes y Análisis
#IVA (19%)
df['IVA'] = df['Total_Venta'] * 0.19

#Ganancia estimada (25%)
df['Ganancia_Estimada'] = df['Total_Venta'] * 0.25

#Pérdida estimada (0.05%)
df['Perdida_Estimada'] = df['Total_Venta'] * 0.0005

#Porcentaje de ventas respecto al total
ventas_totales = df['Total_Venta'].sum()

df['Porcentaje_Ventas'] = (
    df['Total_Venta'] / ventas_totales
) * 100
#Productos con más del 15%
df[df['Porcentaje_Ventas'] > 15]


#Punto 4 – Estadística Descriptiva

#Media
media = df['Precio'].mean()
print("Media:", media)

#Mediana
mediana = df['Precio'].median()
print("Mediana:", mediana)

#Moda
moda = df['Precio'].mode()
print("Moda:")
print(moda)

#Desviación estándar
desviacion = df['Cantidad'].std()

print("Desviación estándar:", desviacion)

#Máximo y mínimo
maximo = df['Total_Venta'].max()
minimo = df['Total_Venta'].min()

print("Máximo:", maximo)
print("Mínimo:", minimo)

#Explicación de las medidas para recordar sus significado para los que no se tengan memoria

#Media: representa el precio promedio de todos los productos.

#Mediana: es el valor central cuando los precios se ordenan de menor a mayor.

#Moda: es el precio que más se repite.

#Desviación estándar: indica qué tan dispersas están las cantidades respecto al promedio.

#Máximo: la mayor venta obtenida por un producto.

#Mínimo: la menor venta obtenida por un producto.



#Punto 5 – Filtrado

#Productos de Tecnología
tecnologia = df[df['Categoria'] == 'Tecnología']

#tecnologia
#Ordenar por Total Venta
df.sort_values(
    by='Total_Venta',
    ascending=False
)


#Exportamos a Excel
df.to_excel(
    'productos_tecnologia.xlsx',
    index=False
)

print("Archivo exportado correctamente.")


#Descargamos productos
from google.colab import files

files.download('productos_tecnologia.xlsx')


```
